# hares-to-json
notebook for creating hares in the repo that can be used without the grid being present

```python
import numpy as np
import pandas as pd
import json
import corner
```

## start with a solar like:

```python
"""
Load in grid
"""

df = pd.read_hdf('/home/oxs235/datastorage/repos_data/ojscutt/pitchfork/data/bob.h5', key='df') #change if accessing from RDS
```

```python
model = df
model = model[(model['star_age'] >= 4.59) & (model['star_age'] <= 4.61)]
model = model[(model['initial_mass'] >= 0.99) & (model['initial_mass'] <= 1.01)]
model = model[(model['initial_Zinit'] >= 0.012) & (model['initial_Zinit'] <= 0.015)]
model = model[(model['initial_Yinit'] >= 0.22) & (model['initial_Yinit'] <= 0.25)]
model = model[(model['initial_MLT'] >= 1.9) & (model['initial_MLT'] <= 2.1)]

model = model.sample(n=1, random_state=42)

inputs = ['initial_mass', 'initial_Zinit', 'initial_Yinit', 'initial_MLT', 'star_age']
outputs = ['calc_effective_T', 'luminosity', 'star_feh'] + [f'nu_0_{i+1}' for i in range(5,40)]

model_inps = model[inputs]
model_outs = model[outputs]

model
```

```python
figure = corner.corner(df[inputs].sample(n=10000), hist_kwargs={'density':True}, truths=model_inps.values.flatten());
```

```python
model.to_json('solar_hare2.json')
```
